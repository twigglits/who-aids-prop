```python
import pysimpactcyan
from psimapacthelper import *
simpact = pysimpactcyan.PySimpactCyan()
```

```python
cfg = { "population.nummen": 500,
        "population.numwomen": 500,
        "population.simtime":20,
        "diagnosis.baseline": -99999,
        "monitoring.cd4.threshold":1,
        "hivtransmission.param.a":-1.3,
        "hivtransmission.param.b":-90,
        "hivtransmission.param.c":0.5,
        "conception.alpha_base": -2.9,
        "dissolution.alpha_0": -2.50494307,
        "population.agedistfile": "/home/jupyter/who-aids-prop/build/python/eswatini_1980.csv",
        "periodiclogging.interval": 0.25
        }

art_intro = {
        "time": 10, #around 2000
        "diagnosis.baseline": -1.75,
        "monitoring.cd4.threshold": 500
    }

baseline_results = simpact.run(
        config=cfg,
        destDir='data',
        seed=1234,
        quiet=True,
        identifierFormat='baseline'
    )

art_results = simpact.run(
        config=cfg,
        interventionConfig=[art_intro],
        destDir='data',
        seed=1234,
        quiet=True,
        identifierFormat='art'
    )
```

```python
baseline_datalist = readthedata(baseline_results)
art_datalist = readthedata(art_results)
```

```python
evaluation_timepoints_inc = np.arange(0, 19.0, 1.0) 
```

```python
baseline_dfs = []
for timepoint in evaluation_timepoints_inc:
        timewindow = [timepoint, timepoint+1]
        result_df = incidence_calculator(datalist=baseline_datalist, agegroup=[15,50],timewindow=timewindow)
        result_df['timepoint'] = timepoint

        # Append the resulting dataframe to the list
        baseline_dfs.append(result_df)

baseline_dfs = pd.concat(baseline_dfs, ignore_index=True)
incidence_wide_df = baseline_dfs.pivot(index='timepoint', columns='Gender', values='incidence').reset_index()
incidence_wide_df.columns.name = None  # Remove the name of the columns index
incidence_wide_df
```

```python
art_dfs = []
for timepoint in evaluation_timepoints_inc:
        timewindow = [timepoint, timepoint+1]
        result_df = incidence_calculator(datalist=art_datalist, agegroup=[15,50],timewindow=timewindow)
        result_df['timepoint'] = timepoint

        # Append the resulting dataframe to the list
        art_dfs.append(result_df)

art_dfs = pd.concat(art_dfs, ignore_index=True)
incidence_wide_df_art = art_dfs.pivot(index='timepoint', columns='Gender', values='incidence').reset_index()
incidence_wide_df_art.columns.name = None  # Remove the name of the columns index
incidence_wide_df_art
```
